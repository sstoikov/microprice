```python
# Connect with IB
# Please reach out to me directly if you need any detailed setup for IB API
ib = IB()
ib.connect('127.0.0.1', 7496, clientId=13)
```

```python
df0 = pd.read_csv('SPY_20190101_.csv', index_col=0)
df0
```

```python
# Define start time at any date but specify as shown below, accurate to second
# Leave end time as empty if you do not want to end at some point

start = datetime.datetime(2019, 1, 7, 15, 45, 26, tzinfo=datetime.timezone.utc)
# start = df.iloc[-1].time + datetime.timedelta(0,1)
end = ""
# start_time = time.time()
df = pd.DataFrame()

while start.replace(tzinfo=None) < datetime.datetime(2019, 1, 10):
    tickers = ib.reqHistoricalTicks(Stock('SPY', 'SMART', 'USD'), start, end, 1000, 'BID_ASK', useRth=False)
    for t in tickers:
        to_append = [t.time.isoformat(' ', 'seconds'), t.sizeBid, t.priceBid, t.sizeAsk, t.priceAsk]
        to_series = pd.Series(to_append)
        df = df.append(to_series, ignore_index=True, sort=False)
    start = tickers[-1].time + datetime.timedelta(0,1)
```
